# 📘 파이썬 기초: 07. 딕셔너리

## 1. 딕셔너리의 개요
* **개념**: 리스트와 마찬가지로 여러 개의 데이터를 저장하는 자료형으로, **키(Key)**와 **값(Value)**의 쌍으로 이루어져 있습니다.
* **리스트와의 차이점**: 리스트는 인덱스(번호)를 사용하여 항목에 접근하지만, 딕셔너리는 사용자가 정의한 '키'를 사용하여 데이터(값)에 접근합니다.
* **생성 방법**: 중괄호 `{ }`를 사용하며, 각 항목은 `키:값` 형식으로 작성하고 쉼표(`,`)로 구분합니다.
* **키의 조건**: 딕셔너리 내에서 키는 중복될 수 없으며 유일해야 합니다. 만약 동일한 키가 여러 개 있다면 마지막 항목만 유효합니다.

```python
prices= {'아메리카노': 2000, '라떼':3000, '아메리카노':1900,'녹차':2500}
```

```python
print(prices)
```

## 2. 딕셔너리 다루기
데이터를 조회, 추가, 삭제하는 방법입니다.
* **값에 접근**: 딕셔너리 이름 뒤 대괄호 `[ ]` 안에 키 이름을 넣어 값을 가져옵니다.
* **항목 추가 및 변경**: 새로운 키에 값을 대입하면 항목이 추가되고, 기존에 있는 키에 값을 대입하면 해당 값이 수정됩니다.
* **항목 삭제 (del)**: 특정 키를 지정하여 해당 항목(키와 값의 쌍)을 삭제합니다.
* **모든 항목 삭제 (clear)**: 딕셔너리 내부의 모든 데이터를 한꺼번에 비웁니다.

```python
# 항목 삭제: pop()은 지정한 위치의 항목을 삭제하고,삭제되는 항목의 값을 반환
a = prices.pop('아메리카노')
print(a)
print(prices)
```

## 3. 딕셔너리 전용 함수
데이터를 효율적으로 관리하기 위한 기능들입니다.
* **키 목록 가져오기 (keys)**: 딕셔너리에 포함된 모든 키들을 모아서 보여줍니다.
* **값 목록 가져오기 (values)**: 딕셔너리에 포함된 모든 값들을 모아서 보여줍니다.
* **쌍으로 가져오기 (items)**: 키와 값의 쌍을 튜플 형태로 묶어서 반환하며, 반복문과 함께 자주 사용됩니다.
* **키 존재 확인 (in)**: 특정 키가 딕셔너리에 들어있는지 확인하여 참/거짓을 반환합니다.

```python
for key in prices.keys():
  print(key)
```

```python
for value in prices.values():
  print(value)
```

```python
#국가 이름 입력 받고 국가의 수도를 알려주는 식
country_capital = {'한국':'서울', '일본':'도쿄', '중국':'베이징'}
country = input('국가 이름을 입력하세요:')
if country in country_capital:
  print(country_capital[country])
else:
  print(f'{country}는 등록되지 않았습니다')
```

### 📊 프로그래밍 언어 선호도 조사
1. **데이터 수집**: 사용자로부터 사용 가능한 프로그래밍 언어를 여러 번 입력받습니다.
2. **카운팅 로직**:
    * 입력받은 언어가 이미 딕셔너리의 키로 존재하면, 해당 키의 값을 1 증가시킵니다.
    * 존재하지 않는 새로운 언어라면, 해당 언어를 키로 추가하고 값에 1을 저장합니다.
3. **결과 출력**: 각 언어별로 몇 번씩 입력되었는지 집계된 데이터를 보여줍니다.

```python
langs={'Python':0, 'C':0, 'Java':0, 'JavaScript':0}

while True:
  s = input('선호하는 언어(Python, C, Java, JavaScript)또는 종료: ')
  if s == '종료':
    print('종료합니다')
    break
  elif s in langs:
    langs[s] += 1
  else:
    print(f'{s}는 등록되지 않은 언어입니다')

for key, value in langs.items():
  print(f'{key}: {value}')
```
