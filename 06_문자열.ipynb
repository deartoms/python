```python
# ëŒ€ì†Œë¬¸ì ë³€í™˜í•˜ê¸°
# ì…ë ¥ë°›ì€ ë¬¸ìì—´ì—ì„œ í•œ ë¬¸ìì”© ëŒ€ë¬¸ìë©´ ì†Œë¬¸ìë¡œ ë³€í™˜, ì†Œë¬¸ìë©´ ëŒ€ë¬¸ìë¡œ ë³€í™˜
s1 = input('ë¬¸ìì—´; ')
s2 =''
for i in s1:
  if i.islower():
    s2 += i.upper()
  elif i.isupper():
    s2 += i.lower()
  else:
    s2 += i
print(s2)
```

```python
# lst = [] ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ í•˜ë‚˜ ë§Œë“¤ê³ 
# whileë¬¸ ì‚¬ìš©
# ì¢‹ì•„í•˜ëŠ” ê²ƒì„ ì…ë ¥ë°›ê³  ('ì¢‹ì•„í•˜ëŠ” ê²ƒì„ ì…ë ¥í•˜ì„¸ìš”: ê°•ì•„ì§€, ê³ ì–‘ì´, í† ë¼, ì¢…ë£Œ')
# 'ì¢…ë£Œ' ì…ë ¥ë˜ë©´ ë°˜ë³µë¬¸ì„ ì¢…ë£Œ
# ì…ë ¥ë°›ì€ ë¬¸ìì—´ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜ .split()
# ë¹ˆ ë¦¬ìŠ¤íŠ¸ì— ë¦¬ìŠ¤íŠ¸ ì¶”ê°€ .extand()
# lst ë¦¬ìŠ¤íŠ¸ ì¶œë ¥
```

```python
lst = []

while True:
    user = input('ì¢‹ì•„í•˜ëŠ” ê²ƒì„ ì…ë ¥í•˜ì„¸ìš”: ê°•ì•„ì§€, ê³ ì–‘ì´, í† ë¼, ì¢…ë£Œ: ')

    if user == 'ì¢…ë£Œ':
        break

    items = user.split()
    lst.extend(items)

print(lst)
```

1. **ë°ì´í„° êµ¬ì¡°**: ì•ŒíŒŒë²³ ë¦¬ìŠ¤íŠ¸(`alphabet`)ì™€ ê·¸ì— ëŒ€ì‘í•˜ëŠ” ëª¨ìŠ¤ ë¶€í˜¸ ë¦¬ìŠ¤íŠ¸(`morse`)ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

    * ì…ë ¥ë°›ì€ ë¬¸ì¥ì„ í•œ ê¸€ìì”© ë¶„ë¦¬í•©ë‹ˆë‹¤.
    * `alphabet`ì—ì„œ í•´ë‹¹ ê¸€ìì˜ ì¸ë±ìŠ¤ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
    * ë™ì¼í•œ ì¸ë±ìŠ¤ë¥¼ ê°€ì§„ `morse`ì˜ ìš”ì†Œë¥¼ ì—°ê²°í•˜ì—¬ ì¶œë ¥í•©ë‹ˆë‹¤.
3. **ëª¨ìŠ¤ ë¶€í˜¸ â†’ ë¬¸ì¥**:
    * ì…ë ¥ë°›ì€ ëª¨ìŠ¤ ë¶€í˜¸ë¥¼ ê³µë°± ê¸°ì¤€ìœ¼ë¡œ `split()`í•©ë‹ˆë‹¤.
    * `morse` ë¦¬ìŠ¤íŠ¸ì—ì„œ ìœ„ì¹˜ë¥¼ ì°¾ì•„ í•´ë‹¹í•˜ëŠ” ì•ŒíŒŒë²³ì„ ì¡°í•©í•©ë‹ˆë‹¤.


```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
morse = ['.-','-...','-.-.','-..','.','..-.','--.',
         '....','..','.---','-.-','.-..','--','-.',
         '---','.--.','--.-','.-.','...','-','..-',
         '...-','.--','-..-','-.--','--..']
```

```python
s = input('ì•ŒíŒŒë²³ ë¬¸ìì—´: ')

for i in range(len(s)):
  if s[i] in alphabet:
    print(morse[alphabet.index(s[i])], end = " ")
```
