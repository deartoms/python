### 튜플 tuple :

* (a, b) 또는 괄호 없이 콤마로 구분
* 파이썬 내부 동작에서 주로 사용, 쉼표를 튜플 생성자로 사용하니 괄호는 생략 가능
* (a, ) 하나의 요소만 갖는 튜플 선언시 요소 뒤에 콤마
* 리스트와 비슷하지만 튜플은 값의 수정 불가능, 불변
* 튜플 자체는 수정이 불가능하지만, 원소로 수정 가능한 리스트나 딕셔너리를 담는 것은 가능


```python
nums = (1, 2, 3, 4)
print(nums)
```

```python
nums = 1,2,3,4
print(nums)
```

```python
nums = (1, )
print(nums)
```

```python
words = "apple", "banana", "grape"
print(type(words))
print(words)
```

```python
nums = (10, 20, 30)
print(nums[0])
print(nums[1:])
```

```python
nums = (5, 7, 4)
print(len(nums))
print(sum(nums))
print(max(nums), min(nums))
print(sorted(nums))
print(nums + nums)
print(nums * 3)
```

```python
nums = (5, 7, 4)
nums += (3,)
print(nums)
```

### 집합 set
* 집합의 원소는 반드시 '값이 변하지 않는(Hashable)' 상태여야 한다. 리스트나 딕셔너리는 언제든 값이 바뀔 수 있기 때문에 집합의 원소가 될 수 없다
* 중복되지 않은 요소들로 구성된 자료형
* 집합내의 모든 요소들은 순서가 없다
* 인덱싱이나 슬라이싱으로 집합 내의 요소에 접근할수 없다
* set1 = {1,2,3}
* set1 = {} 는 딕셔너리로 인식
* set2 = set()
* 연산자 : | 합집합, & 교집합, - 차집합, ^ 대칭 차집합(합집합에서 교집합을 뺀 집합)
* 내장함수 : add(), remove(), update()


```python
set1 = {1,2,3,4,5}
set2 = {4, 5, 6}
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)
print(set1 ^ set2)
print(1 in set1)
print(len(set1))
```

```python
two_mul = {2, 4, 6, 8, 10, 12}
three_mul = {3, 6, 9, 12}
print(two_mul & three_mul)
print(2 in two_mul)
print({3, 6} in three_mul)
```

```python
nums = {1,2,3}
nums.add(4)
nums.add(2)
print(nums)
```

```python
nums = {1, 2, 3}
nums.remove(2)
print(nums)
nums.remove(4)
```

```python
nums = {1, 2, 3}
nums.update([3, 4, 5])
print(nums)
```

```python
two_mul = {2, 4, 6}
two_mul.add(8)
print(two_mul)
two_mul.remove(8)
print(two_mul)
two_mul.update({8, 10, 12})
print(two_mul)
```
