### ë¬¸ì œ: ê²½ì§„ëŒ€íšŒ í‰ê°€ ì ìˆ˜ êµ¬í•˜ê¸°

*   ì‹¬ì‚¬ìœ„ì› judge_num ì´ 5ëª…ì´ ì‹¬ì‚¬
*   score ê° ì ìˆ˜ë¥¼ ì…ë ¥ ë°›ê³ 
*   ì ìˆ˜ ì´ í•©ê³„ sumì˜ í‰ê· ì„ ì¶œë ¥
*   ìµœê³  ì ìˆ˜ëŠ” ë¹¼ê³  í•©ê²Œë¥¼ êµ¬í•´ì•¼ í•¨ max_score
*   í˜„ì¬ ì ìˆ˜ê°€ ìµœê³  ì ìˆ˜ë³´ë‹¤ í¬ë©´, í˜„ì¬ ì ìˆ˜ë¥¼ max_scoreì— ì €ì¥í•˜ê³  sumì—ì„œ ë¹¼ì¤˜ì•¼ í•©ë‹ˆë‹¤



```python
judge_num = 5
sum = 0
max_score = 0

for i in range(judge_num): #0,1,2,3,4
    score = int(input('ì ìˆ˜: ')) # 100, 80 , 70, 90, 100
    sum += score # 100+80+70+90+100

    if score > max_score:
        max_score = score # 100
total = sum
sum -= max_score
average = sum / (judge_num - 1)

print(f'ì ìˆ˜ ì´ í•©ê³„: {total}ì ')
print(f'ìµœê³  ì ìˆ˜ ì œì™¸ í‰ê· : {average}ì ')
```

    * `random.randint(1, 100)`ìœ¼ë¡œ ì •ë‹µì„ ìƒì„±í•©ë‹ˆë‹¤.
    * ì‚¬ìš©ìê°€ ì •ë‹µì„ ë§í ë•Œê¹Œì§€ `while` ë¬¸ìœ¼ë¡œ ì…ë ¥ì„ ë°˜ë³µí•©ë‹ˆë‹¤.
    * ì…ë ¥ê°’ì´ í¬ë©´ "ë„ˆë¬´ ë†’ì•„ìš”", ì‘ìœ¼ë©´ "ë„ˆë¬´ ë‚®ì•„ìš”"ë¥¼ ì¶œë ¥í•˜ì—¬ ë²”ìœ„ë¥¼ ì¢í˜€ê°‘ë‹ˆë‹¤.
2. **ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°**:
    * ì‚¬ìš©ìê°€ ìˆ«ìë¥¼ í•˜ë‚˜ ì •í•˜ë©´ ì»´í“¨í„°ê°€ `low`ì™€ `high` ë³€ìˆ˜ë¥¼ ì¡°ì •í•˜ë©° ì •ë‹µì„ ì°¾ì•„ê°€ëŠ” ë¡œì§ì…ë‹ˆë‹¤.

```python
# 1. ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random

print('1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤.')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”!')

answer = random.randint(1, 100) # 1~100ì‚¬ì´ ìˆ«ì í•˜ë‚˜ ëœë¤ìœ¼ë¡œ ì €ì¥
times = 0 # ë„ì „ íšŸìˆ˜
# while ë¬¸ìœ¼ë¡œ ë°˜ë³µí•˜ëŠ”ë°. ì¦‰ ë„ì „ì´ ê³„ì†
while True:
  user = int(input('1~100 ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”'))
  times += 1 # ë„ì „ íšŸìˆ˜ 1 ì¦ê°€
# userê°€ ì…ë ¥í•œ ìˆ«ìê°€ ë§ìœ¼ë©´ ë°˜ë³µë¬¸ ì¢…ë£Œê°€ ë˜ê³ 
  if user == answer:
    break
# userê°€ ì…ë ¥í•œ ìˆ«ì í¬ë©´ í¬ë‹¤ê³  ì¶œë ¥
  elif user > answer:
    print(f'{user}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤')
# userê°€ ì…ë ¥í•œ ìˆ«ì ì‘ìœ¼ë©´ ì‘ë‹¤ê³  ì¶œë ¥
  else:
    print(f'{user}ë³´ë‹¤ í½ë‹ˆë‹¤')
# ìµœì¢… ë„ì „ ë° í™”ë©´ì— ì •ë‹µì„ ë§ì·„ëŠ”ì§€ ì¶œë ¥
print(f'ì •ë‹µì…ë‹ˆë‹¤! {times}íšŒ ë§Œì— ë§ì·„ìŠµë‹ˆë‹¤')
```

```python
# 1. ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random

print('1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤.')
print('ì´ ìˆ«ìë¥¼ ë§ì¶°ë³´ì„¸ìš”!')

answer = random.randint(1, 100)
time = 0

while True:
    user = int(input('ìˆ«ì ì…ë ¥: '))
    time += 1

    if user == answer:
        print(f'ì •ë‹µì…ë‹ˆë‹¤! ë„ì „ íšŸìˆ˜: {time}ë²ˆ')
        break
    elif user > answer:
        print('ë” ì‘ì€ ìˆ«ìì…ë‹ˆë‹¤.')
    else:
        print('ë” í° ìˆ«ìì…ë‹ˆë‹¤.')
```

```python
# 2.ì‚¬ìš©ìê°€ ìƒê°í•œ ìˆ˜ ë§íˆê¸°
print('1~100 ì‚¬ì´ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í•˜ì„¸ìš”')
print('ì´ ìˆ«ìë¥¼ ë§í˜€ë³´ê² ìŠµë‹ˆë‹¤!')

low = 1
high = 100
times = 0

while True:
  print(f'low: {low}, high: {high}')
  num = (low+high)//2 #ì»´í“¨í„°ê°€ ì œì‹œí•œ ìˆ«ì
  answer = input(f'{num}ì¸ê°€ìš”? ok / up / down') # ì‚¬ìš©ìê°€ ë‹µë³€ì„ í•´ì¤Œ
  times += 1

  if answer == 'ok':
    break
  elif answer == 'up': # numë³´ë‹¤ í¬ë‹¤ê³  í•´ì„œ lowê°€ num+1
    low = num + 1
  elif answer == 'down': # numë³´ë‹¤ ì‘ë‹¤ê³  í•´ì„œ highê°€ num-1
    high = num -1
  else:
    print('ì˜ëª» ì…ë ¥í–ˆìŠµë‹ˆë‹¤. ')
print(f'{times}íšŒ ë§Œì— ë§ì·„ìŠµë‹ˆë‹¤.')
```
