````python
import math
# ì‚¬ê°í˜•square, ì‚¼ê°í˜•triangle, ì›circle í•¨ìˆ˜ 3ê°œ
# ì‚¬ê°í˜•ì€ ê°€ë¡œwidth ì„¸ë¡œheight ì…ë ¥ ë°›ì•„ì„œ ì‚¬ê°í˜• ë„“ì´ëŠ” ê°€ë¡œ*ì„¸ë¡œ ê°’ ë°˜í™˜

def square(width, height):
    return width * height
# ì‚¼ê°í˜•ì€ ë°‘ë³€base ë†’ì´height ì…ë ¥ ë°›ì•„ì„œ ì‚¼ê°í˜• ë„“ì´ëŠ” ë°‘ë³€*ë†’ì´/2 ê°’ ë°˜í™˜

def triangle(base, height1):
    return base * height1 / 2
# ì›ì€ ë°˜ì§€ë¦„radiusë¥¼ ì…ë ¥ ë°›ì•„ì„œ ì› ë„“ì´ëŠ” math.pi*ë°˜ì§€ë¦„*ë°˜ì§€ë¦„ ê°’ ë°˜í™˜ (import math ë¼ì´ë¸ŒëŸ¬ë¦¬ ì‚¬ìš©)

def circle(radius):
    return math.pi * radius * radius

# ë„í˜•ì˜ ì¢…ë¥˜ë¥¼ ì…ë ¥ ë°›ì•„ì„œ, ë³€ìˆ˜ shape_type1, shape_type2
# ì‚¬ê°í˜•ì´ë©´ ì‚¬ê°í˜• í•¨ìˆ˜ í˜¸ì¶œ, ì‚¼ê°í˜•ì´ë©´ ì‚¼ê°í˜• í•¨ìˆ˜ í˜¸ì¶œ, ì›ì´ë©´ ì› í•¨ìˆ˜ í˜¸ì¶œ
# area ë³€ìˆ˜ì—ëŠ” ê° ë„í˜•ì˜ ë„“ì´ ê°’ì´ ì…ë ¥ë¨

shape_type1 = input('ì²« ë²ˆì§¸ ë„í˜• ì…ë ¥(ì‚¬ê°í˜•/ì‚¼ê°í˜•/ì›): ')
if shape_type1 == "ì‚¬ê°í˜•":
    width = int(input('ê°€ë¡œ: '))
    height = int(input('ì„¸ë¡œ: '))
    area1 = square(width, height)
elif shape_type1 == "ì‚¼ê°í˜•":
    base = int(input('ë°‘ë³€: '))
    height1 = int(input('ë†’ì´: '))
    area1 = triangle(base, height1)
else:
    radius = int(input('ë°˜ì§€ë¦„: '))
    area1 = circle(radius)


shape_type2 = input("ë‘ ë²ˆì§¸ ë„í˜• ì…ë ¥(ì‚¬ê°í˜•/ì‚¼ê°í˜•/ì›): ")
if shape_type2 == "ì‚¬ê°í˜•":
    width = int(input('ê°€ë¡œ: '))
    height = int(input('ì„¸ë¡œ: '))
    area2 = square(width, height)
elif shape_type2 == "ì‚¼ê°í˜•":
    base = int(input('ë°‘ë³€: '))
    height1 = int(input('ë†’ì´: '))
    area2 = triangle(base, height1)
else:
    radius = int(input('ë°˜ì§€ë¦„: '))
    area2 = circle(radius)

# area ë³€ìˆ˜ì— ì…ë ¥ëœ ê°’ area1ê³¼ area2ë¥¼ ë¹„êµí•´ì„œ ì–´ë–¤ ë„í˜•ì´ ì–¼ë§ˆë§Œí¼ ë‹¤ ë„“ì€ì§€ ì¶œë ¥. ë„í˜•ì´ í¬ê¸°ê°€ ê°™ìœ¼ë©´ ê°™ë‹¤ê³  ì¶œë ¥

if area1 > area2:
    print("ì²« ë²ˆì§¸ ë„í˜•ì´ ë” ë„“ìŒ")
elif area1 < area2:
    print("ë‘ ë²ˆì§¸ ë„í˜•ì´ ë” ë„“ìŒ")
else:
    print("ë‘ ë„í˜•ì˜ ë„“ì´ê°€ ê°™ìŒ")
````
