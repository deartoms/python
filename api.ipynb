```python
import requests # HTTP 요청을 보내기 위한 requests 라이브러리
import pandas as pd
import pprint
```

```python
url = "https://www.omdbapi.com/"
params = { # API 요청에 필요한 파라미터들을 딕셔너리로 정의
    "apikey": "1558c3fe",
    "t": "Avengers",
    "type": "movie",
    "y": "2012"
}
# GET 요청
response = requests.get(url, params=params)
print(response.status_code) # HTTP 응답 상태 코드(예: 200은 성공)를 출력
```

```python
# JSON 응답을 파이썬 객체로 변환
data = response.json() # 응답 본문을 JSON 형식으로 파싱하여 파이썬 딕셔너리로 변환
pprint.pprint(data)
```

```python
import requests

url = "https://www.omdbapi.com/"
params = {
    "apikey": "1558c3fe",
    "s": "Avengers",# t 대신 s 사용 -> 제목에 "Avengers" 포함된 영화 전부 검색
    "type": "movie"
}

response = requests.get(url, params=params)
data = response.json()
data
```

```python
# 검색 결과 리스트 출력
movies = data["Search"]
for movie in movies:
  print(movie["Title"], movie['Year'])
```

```python
import pandas as pd

if 'Search' in data:
  m_df = pd.DataFrame(data['Search'])
else:
  m_df = pd.DataFrame([data]) # 'Search' 키가 없으면, data 자체가 단일 객체이므로 리스트로 감싸서 DataFrame으로 만듭니다.

display(m_df.head())
```

```python
m_df = m_df.iloc[0:4]# 인덱스 0부터 3까지의 행만 선택합니다.
display(m_df)
```

```python
# 어벤져스 시리즈 4편 정보 수집

# m_df의 'Title'과 'Year' 컬럼 값으로 titles 리스트를 새엇ㅇ
titles = [(row['Title'], row['Year']) for index, row in m_df.iterrows()]
print(titles)
# 각 영화 정보를 저장할 빈 리스트
movies = []

url = "https://www.omdbapi.com/"
for title, year in titles:
  params = {
    "apikey": "1558c3fe",
    "t": title,
    "type": "movie",
    "y": year
  }
  response = requests.get(url, params=params) # RESTful api에서 데이터 요청 get 방식
  data = response.json()
  movies.append({
      "title": data.get('Title'), #get()메서드를 사용하면 'Title' 키가 딕셔너리에 없을 경우 오류 대신 None
      "year": data.get('Year'),
      "director": data.get('Director'),
      "imdbRating": data.get('ImdbRating'),
      "genre": data.get('Genre')
  })

  pprint.pprint(movies)

  # 데이터프레임으로 변환
  df = pd.DataFrame(movies)
  display(df)
```
