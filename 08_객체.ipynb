# 📘 파이썬 기초: 10. 객체

## 1. 객체와 클래스의 이해
* **객체 지향 프로그래밍(Object Oriented Programming)**: 현실 세계의 모든 사물을 '객체'로 보고, 이를 프로그램으로 구현하는 기법입니다.
* **객체(Object)**: 현실 세계에 존재하는 모든 것이 객체가 될 수 있으며, **속성**과 **메서드**로 구성됩니다.
    * **속성(Attribute)**: 객체의 상태나 특징을 나타내는 변수입니다.
    * **메서드(Method)**: 객체의 동작이나 기능을 나타내는 함수입니다.
* **클래스(Class)**: 객체를 만들기 위한 '설계도' 또는 '틀'입니다. 클래스를 통해 생성된 실제 사물을 '인스턴스(Instance)' 또는 '객체'라고 부릅니다.

## 2. 클래스 생성 및 사용
* **클래스 정의**: `class` 키워드를 사용하며, 내부에 속성(변수)과 메서드(함수)를 정의합니다.
* **생성자 (`__init__`)**: 객체가 생성될 때 자동으로 호출되는 특별한 메서드로, 주로 객체의 초기 상태(속성)를 설정하는 데 사용됩니다.
* **Self 키워드**: 메서드 내부에서 객체 자신의 속성에 접근할 때 사용하며, 모든 메서드의 첫 번째 매개변수로 지정해야 합니다.
* **정보 은닉 (Private)**: 속성 이름 앞에 `__`(언더바 두 개)를 붙여 클래스 외부에서 직접 수정하지 못하도록 보호할 수 있습니다.

```python
class Student:
  def __init__(self, name, math, computer):
    self.__name = name
    self.__math = math
    self.__computer = computer

  def get_name(self):
    return self.__name

  def get_average(self):
    return (self.__math + self.__computer) / 2

  def set_math(self, math):
    self.__math = math

  def set_computer(self,computer):
    self.__computer = computer

s1 = Student('kim',95,89)
s2 = Student('lee',88,75)
#print(s1.__name)
print(s1.get_name())
print(s1.get_average())

s1.set_computer(97)
print(s1.get_average())
```

## 3. 상속 (Inheritance)
* **개념**: 기존 클래스(부모 클래스)의 속성과 메서드를 그대로 물려받아 새로운 클래스(자식 클래스)를 만드는 기능입니다.
* **장점**: 코드의 재사용성을 높여주며, 중복된 코드를 줄여 유지보수가 쉬워집니다.
* **메서드 오버라이딩(Overriding)**: 부모 클래스로부터 물려받은 메서드를 자식 클래스에서 용도에 맞게 새롭게 재정의하여 사용하는 것을 말합니다.

```python
# 부모 클래스
class Student:
    def __init__(self, name, computer):
        self.name = name
        self.computer = computer

    def set_computer(self, computer):
        self.computer = computer

    def get_name(self):
        return self.name
```

```python
# 자식 클래스
class Social(Student):
  def __init__(self, name, computer, social):
    super().__init__(name, computer)
    self.social = social

  def set_social(self, social):
    self.social = social

  def get_average(self):
    return (self.social + self.computer) / 2
```

```python
# 자식 클래스
class Science(Student):
  def __init__(self, name, computer, science):
    super().__init__(name, computer)
    self.science = science

  def set_science(self, science):
    self.science = science

  def get_average(self):
    return (self.science + self.computer) / 2
```

```python
st1 = Science('lee', 90, 80)
print(st1.get_name())
print(st1.get_average())
```

```python
st2 = Social('kim', 98, 88)
print(st2.get_name())
print(st2.get_average())
st2.set_computer(100)
print(st2.get_average())
```

```python
# 부모 클래스
class Figure:
  def __init__(self, area, perimeter):
    self.area = area
    self.perimeter = perimeter

  def get_area(self):
    return self.area

  def get_perimeter(self):
    return self.perimeter
```

```python
#자식 클래스
class Circle(Figure):
  def __init__(self, radius):
    area = 3.14*radius**2
    perimeter = 2*3.14*radius
    super().__init__(area, perimeter)
    self.__radius = radius

  def set_radius(self, radius):
    self.radius = radius
    self.area = 3.14*radius**2
    self.perimeter = 2*3.14*radius

  def get_radius(self):
    return self.radius
```

```python
#자식 클래스
class Rectangle(Figure):
  def __init__(self, width=10, height=10):
    area = width*height
    perimeter = 2*(width+height)
    super().__init__(area, perimeter)
    self.width = width
    self.height = height

  def set_width(self, width):
    self.width = width
    self.area = width*self.height
    self.perimeter = 2*width+2*self.height

  def set_height(self, height):
    self.height = height
    self.area = self.width*height
    self.perimeter = 2*self.width+2*height

  def get_width(self):
    return self.width

  def get_height(self):
    return self.height
```

```python
f1 = Circle(7)
print(f1.area)
print(f1.perimeter)
print(f1.get_area(),f1.get_perimeter())
f1.set_radius(10)
print(f1.get_area(), f1.get_perimeter())
```

### 🚗 자동차 클래스와 객체
1. **클래스 설계**: `Car` 클래스를 생성하고, 모델명(`__model`)과 연비(`__fuelefficiency`)를 속성으로 정의합니다.
2. **정보 보호**: 속성들을 `private`으로 설정하여 외부의 직접 접근을 막고, 안전하게 값을 읽거나 수정할 수 있는 전용 메서드(`get_model`, `set_fuelefficiency` 등)를 만듭니다.
3. **연료량 계산 로직**: 주행 거리를 입력받아 (주행 거리 / 연비) 공식을 통해 필요한 연료량을 계산하여 반환하는 메서드를 구현합니다.
4. **객체 생성**: 정의된 클래스를 바탕으로 `car1`, `car2` 등 서로 다른 모델명과 연비를 가진 실제 자동차 객체들을 만들어 실행합니다.

```python
# 'GV70', 12
# distance = input(주행거리 : )

class Car:
  def __init__(self, model, fuelefficiency): # car1 = Car('GV70', 12)
    self.__model = model
    self.__fuelefficiency = fuelefficiency

  def get_model(self):
    return self.__model

  def get_fuelefficiency(self):
    return self.__fuelefficiency

  def set_fuelefficiency(self, fuelefficiency):
    self.__fuelefficiency = fuelefficiency

  def get_fuelamount(self, distance):
    return distance / self.__fuelefficiency
```

```python
car1 = Car('GV70', 12)
print(f'모델명: {car1.get_model()}')
print(f'연비: {car1.get_fuelefficiency()}')
car1.set_fuelefficiency(15)
distance = float(input('주행 거리(km): '))
print(f'연료량: {car1.get_fuelamount(distance)}')
```
